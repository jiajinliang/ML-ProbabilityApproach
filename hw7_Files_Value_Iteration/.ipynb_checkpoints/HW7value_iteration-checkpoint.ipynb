```python
import numpy as np
import math

west_trasition_matrix = np.genfromtxt("./prob_a1.txt");
p_a2 = np.genfromtxt("'hw7_prob_a2.txt');
p_a3 = np.genfromtxt('hw7_prob_a3.txt');
p_a4 = np.genfromtxt('hw7_prob_a4.txt');
rewards = np.genfromtxt('hw7_rewards.txt');
S = 81        # state 
A = 4         # action
gamma = 0.99  
"
print (west_trasition_matrix.shape, p_a2.shape, p_a3.shape, p_a4.shape, rewards.shape)

# transition from given datas
p_a = []
p_a.append(west_trasition_matrix)
p_a.append(p_a2)
p_a.append(p_a3)
p_a.append(p_a4)

# action --> w/n/e/s  4 possible.
A_list = []
A_list.append("West")  # West
A_list.append("North") # North
A_list.append("East")  # East
A_list.append("South") # South

#transition 
tr_a1 = []
tr_a2 = []
tr_a3 = []
tr_a4 = []
transition_list = []

#init
for i in range(S*S):
    tr_a1.append(0.0)
    tr_a2.append(0.0)
    tr_a3.append(0.0)
    tr_a4.append(0.0)

#print (tr_a1.shape,tr_a2.shape,tr_a3.shape,tr_a4.shape)
#-----------------------------
tr_a1=np.array(tr_a1)
tr_a2=np.array(tr_a2)
tr_a3=np.array(tr_a3)
tr_a4=np.array(tr_a4)

#print (tr_a1.shape,tr_a2.shape,tr_a3.shape,tr_a4.shape)

tr_a1=np.reshape(tr_a1,(S,S))
tr_a2=np.reshape(tr_a2,(S,S))
tr_a3=np.reshape(tr_a3,(S,S))
tr_a4=np.reshape(tr_a4,(S,S))
#print (tr_a1.shape,tr_a2.shape,tr_a3.shape,tr_a4.shape)

transition_list.append(tr_a1)
transition_list.append(tr_a2)
transition_list.append(tr_a3)
transition_list.append(tr_a4)

for i in range(A):
    
    p_t = p_a[i]
    t_m = transition_list[i]
    
    for p in p_t:
        
        p_i1=int(p[0])
        p_i2=int(p[1])
        t_m[p_i1-1][p_i2-1] = p[2]
```

```python
def transition_matrix(s,sp,a):
    
    s_i=int(s)
    sp_i=int(sp)
    return transition_list[a][s_i][sp_i]
```

```python
# Initialize
v_temp = []
for i in range(S):
    v_temp.append (0.0)
```

```python
def VP():
    
    for s in range(S):
        temp = [] 
        
        for a in range(A): 
            su = 0.0

            for sp in range(S):
                su += 1.0 * transition_matrix(s,sp,a) * v_temp[sp]
                
            temp.append(su)
            
        v_temp[s] = rewards[s] + gamma * max(temp)
```

```python
# (a)  V*(S)
t=1000
for i in range(t): #1000 times
    VP()
    
print (v_temp)
```

```python
for i in range(t):#1000+1000 times
    VP()
    
print (v_temp)
```

```python
#after value iteration, prints values.
v_l=len(v_temp)
for i in range(v_l):
    print (v_temp[i])
```

```python
#(b) optimal policy pi*(s)
pi_p = []

for s in range(S):
    li_q = []
    
    for a in range(A):
        sum_temp = 0
        
        for sp in range(S):
            sum_temp += 1.0 * transition_matrix(s,sp,a) * v_temp[sp]
            
        li_q.append(rewards[s] + 1.0 * sum_temp* gamma )
        
    pi_p.append(A_list[int(li_q.index(max(li_q)))])
```

```python
for i in range(S):
    print (pi_p[i])
```
