```python
import numpy as np
import math
def transition_matrix(s,sp,a):
    s_i=int(s)
    sp_i=int(sp)
    return transition_list[a][s_i][sp_i]

west_trasition_matrix = np.genfromtxt("./prob_a1.txt");
north_trasition_matrix = np.genfromtxt("./prob_a2.txt");
east_trasition_matrix = np.genfromtxt("./prob_a3.txt");
south_trasition_matrix = np.genfromtxt("./prob_a4.txt");
rewards = np.genfromtxt("./rewards.txt");
S = 81  
A = 4 
discount_factor = 0.99  
#print (west_trasition_matrix.shape, north_trasition_matrix.shape, east_trasition_matrix.shape, south_trasition_matrix.shape, rewards.shape)
actions = []
actions.append(west_trasition_matrix)
actions.append(north_trasition_matrix)
actions.append(east_trasition_matrix)
actions.append(south_trasition_matrix)
action_list = []
action_list.append("West")  
action_list.append("North") 
action_list.append("East")  
action_list.append("South") 
transition_west = []
transition_north = []
transition_east = []
transition_south = []
transition_list = []
for i in range(S*S):
    transition_west.append(0.0)
    transition_north.append(0.0)
    transition_east.append(0.0)
    transition_south.append(0.0)
transition_west=np.array(transition_west)
transition_north=np.array(transition_north)
transition_east=np.array(transition_east)
transition_south=np.array(transition_south)

transition_west=np.reshape(transition_west,(S,S))
transition_north=np.reshape(transition_north,(S,S))
transition_east=np.reshape(transition_east,(S,S))
transition_south=np.reshape(transition_south,(S,S))

transition_list.append(transition_west)
transition_list.append(transition_north)
transition_list.append(transition_east)
transition_list.append(transition_south)

for i in range(A):    
    a = actions[i]
    b = transition_list[i]
    for p in a:
        p_i1=int(p[0])
        p_i2=int(p[1])
        b[p_i1-1][p_i2-1] = p[2]
v_temp = []
for i in range(S):
    v_temp.append (0.0)
def findvaluestar():    
    for s in range(S):
        temp = [] 
        for a in range(A): 
            su = 0.0
            for sp in range(S):
                su += 1.0 * transition_matrix(s,sp,a) * v_temp[sp]
            temp.append(su)
        v_temp[s] = rewards[s] + discount_factor * max(temp)
t=1000
for i in range(t): 
    findvaluestar()
print (v_temp)
for i in range(t):
    findvaluestar()
print (v_temp)
value_list=len(v_temp)
for i in range(value_list):
    print (v_temp[i])
pai = []

for s in range(S):
    listofq = []
    for a in range(A):
        temp2 = 0
        for sp in range(S):
            temp2 += 1.0 * v_temp[sp] * transition_matrix(s,sp,a)
        listofq.append(rewards[s] + 1.0 * temp2* discount_factor )
    pai.append(action_list[int(listofq.index(max(listofq)))])
for i in range(S):
    print (pai[i])
```
