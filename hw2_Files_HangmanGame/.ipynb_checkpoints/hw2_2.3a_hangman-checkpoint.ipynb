```python
import re

f = open("./hw2_word_counts_05.txt", 'r')
lines = f.readlines()
f.close()

count=0
for line in lines:
    word= re.split ('[ \n]+', line)
    count=count+int(word[1])
    #print (word[2])
x=[]
y=[]
for line in lines:
    word= re.split ('[ \n]+', line)
    word.insert(2, int(word[1])/count)
    #word[1]=int(word[1])
    
    #print (word[1])
    #alist = sorted(word[0])
    #print (word[0])
    x.insert(0,int(word[1]))
    y.insert(0,word[0])
#print (y)
z = sorted(zip(x,y))
#print (z)
print ("14 least frequent: ",z[0:14])
z = sorted(zip(x,y),reverse=True)
#print (z)
print ("15 most frequent: ", z[0:15])
add = 1
cg = "00000"
ig = "EA"
possible=[]
prob=[]
for line in lines:
    word = re.split('[ \n+]',line)
    word.insert(2, int(word[1])/count)
    #print (word)
    for i in range (0,5):
        if (cg[i] != '0'):
            if(word[0][i] != cg[i]):
                add=0
                break
        elif ((cg.find(word[0][i]) != -1)):
            #print(word[0])
            #print(word[0][i])
            add=0
            break
        if((ig.find(word[0][i]) != -1)):
            #print(word[0])
            add = 0
            break
            #continue
            #print
    if (add == 1):
        possible.insert(0,word[0])
        prob.insert(0,word[1])
        add = 0
    add = 1
#print (word)
print(possible)
print(prob)
for j in prob:
    sum = sum + int(j)
print (sum)
newp = []
for k in prob:
    newp.append(int(k)/sum)
print (newp)

p = []
q = []
for i in range (0,26):
    p.insert(0,0)
    q.insert(0,chr(90-i))
#print (p)
#print (q[25])
for i in range (0,5):
    if(cg[i]=='0'):
        for c in range (65,91):
            for word in possible:
                if(word[i]==chr(c)):
                    p[c-65] = p[c-65] + newp[possible.index(word)]
print (p)
#pmax = max(p)
sum = 0
#print (prob)
#next = q(p.index(pmax))
#print(next)
z = sorted(zip(p,q))
print (z)            
```
